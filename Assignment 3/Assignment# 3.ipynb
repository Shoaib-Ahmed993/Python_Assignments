```python
# 7. Write a program which takes 5 inputs from user for different subjectâ€™s marks, total it
# and generate mark sheet using grades ?

sub1 = float(input("Enter marks of first subject: "));
sub2 = float(input("Enter marks of second subject: "));
sub3 = float(input("Enter marks of third subject: "));
sub4 = float(input("Enter marks of dourth subject: "));
sub5 = float(input("Enter marks of fifth subject: "));
obtained = sub1+sub2+sub3+sub4+sub5;
total = 500;
percentage = (obtained/total)*100;
print("percentage: "+ str(percentage) + "%")
if (percentage < 100 and percentage >= 80):
    print("A+ Grade");
elif (percentage < 80 and percentage >= 70):
    print("A Grade");
elif (percentage < 70 and percentage >= 60):
    print("B Grade");
elif (percentage < 60 and percentage >= 50):
    print("C Grade");
elif (percentage < 50 and percentage >= 40):
    print("D Grade");
elif (percentage < 40 and percentage >=30):
    print("E Grade");
elif (percentage < 30):
    print("Fail");
else: 
    print("You have entered incorrect value");
```
